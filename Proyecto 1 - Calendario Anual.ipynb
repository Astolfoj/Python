### Algoritmo del calculo del día de la semana
Un algoritmo para calcular el día de la semana para una fecha particular se muestra a continuación:

*Determinar el dia de la semana para un mes, dia y año.*
    
    1. Hacer digitos_siglo igual a los dos primeros digitos del año
    2. Hacer digitos_año igual a los dos ultimos digitos del año
    3. Hacer valor igual a digitos_año + floor(digitos_año / 4)
    4. Si digitos_siglo es igual a 18, añadir 2 a valor, de lo contrario
       si digitos_siglo es igual a 20, añadir 6 a valor
    5. Si el mes es igual a Enero y el año no es bisiesto, entonces añadir 1 a valor,
       de lo contrario
       Si el mes es igual a Febrero y el año es bisiesto, entonces añadir 3 a valor;
       si no es año bisiesto, entonces añadir 4 a valor, de lo contrario
       Si el mes es igual a Marzo o Noviembre, añadir 4 a valor, de lo contrario
       Si el mes es igual a Abril o Julio, entonces añadir 0 a valor, de lo contrario
       Si el mes es igual a Mayo, entonces añadir 2 a valor, de lo contrario
       Si el mes es igual a Junio, entonces añadir 5 a valor, de lo contrario
       Si el mes es igual a Agosto, entonces añadir 3 a valor, de lo contrario
       Si el mes es igual a Octubre, entonces añadir 1 a valor, de lo contrario
       Si el mes es igual a Setiembre o Diciembre, añadir 6 a valor
    6. Hacer valor igual a (valor + dia) % 7
    7. Si valor es igual a 1, el dia de la semana será Domingo; de lo contrario
       Si valor es igual a 2, el dia de la semana será Lunes; de lo contrario
       Si valor es igual a 3, el dia de la semana será Martes; de lo contrario
       Si valor es igual a 4, el dia de la semana será Miercoles; de lo contrario
       Si valor es igual a 5, el dia de la semana será Jueves; de lo contrario
       Si valor es igual a 6, el dia de la semana será Viernes; de lo contrario
       Si valor es igual a 0, el dia de la semana será Sabado; de lo contrario
    
También debemos de determinar si el año es bisiesto, basado en el algoritmo de validación del año bisiesto:

    if anio % 4 is 0 and anio % 100 is not 0 or anio % 400 is 0:
        anio_bisiesto = True
    else:
        anio_bisiesto = False
        
Verifiquemos el algoritmo para una fecha: 24 de mayo 2025. La variable `digitos_siglo` será 20 y `digitos_año` será 25. La variable `valor` será asignada con:

    valor = digitos_año + floor(digitos_año) / 4)
          = 25 + floor(25/4) -> 24 + floor(6.25) -> 25 + 6  -> 31
          
Como `digitos_siglo` es igual a 20, `valor` se incrementa por 6:

    valor = valor + 6 -> 31 + 6 = 37
    
Como el mes el igual a Mayo, `valor` se incrementa por 2:

    valor = valor + 2 -> 37 + 2 = 39
    
Se actualiza `valor` según en dia del mes (24):

    valor = (valor + dia_del_mes) % 7
          = (39 + 24) % 7
          = 63 % 7
          = 0
          
De lo que se concluye que el 24 de Mayo de 2025 será Sábado.

### Algoritmo para obtener el calendario del mes
Aqui es donde tiene que construír su propio algoritmo. Es básicamente construír una tabla considerando el inicio del mes y un número de columnas igual a 7.

Debe de utilizar el algoritmo anterior para hallar el dia 1 para un mes, de acuerdo al año ingresado (Recuerde validar el dato ingresado):

    Ingrese el año [1800 - 2099]:
    
Al hallar el dia que le corresponde al 1 del mes podrá saber construír el calendario.

El caso mas sencillo es que el dia del mes, según el algoritmo anterior sea 1 (Domingo). Si el mes ingresado es Enero, será cuestion de colocar los números iniciando desde la primera columna hasta la 7 y luego bajar a la siguiente fila y seguir colocando los números de dias hasta llegar al día 31. Todos estos valores, siempre alineados a la cabecera con los nombres de los días.

Construya poco a poco el programa final hasta obtener un resultado terminado. ¡No se olvide comentar su programa!

Una vez que sus algoritmos son terminados, podrá aplicarlos para los 12 meses del año ingresado.

```python
#Integrantes:
#1.- 
#2.- 
#3.- 
#############################################################################

# ESCRIBA SU PROGRAMA EN ESTA SECCION
#Generando lista de meses y otras cosas

lista_meses = [[["" for i in range(7)] for i in range(6)] for i in range(12)]
traductor = {1:0,2:1,3:2,4:3,5:4,6:5,0:6}
longitud_meses=[31,28,31,30,31,30,31,31,30,31,30,31]
nombres_meses=["Enero","Febrero","Marzo","Abril","Mayo","Junio",
            "Julio","Agosto","Septiembre","Octubre","Noviembre",
            "Diciembre"]

# Pida al usuario el año (valide el ingreso) (a parte de estructuras if, else, use también la estructura try except)
while True:
    try:
        año = input("Ingrese el año[1800-2099]:")
        digitos_siglo=int(año[:2])
        digitos_año=int(año[2:])
        año = int(año)
        if año>=1800 and año<=2099:
            break
        else:
            print("El año debe estar en el rango 1800-2099")
    except:
        print("El año debe ser un numero entero")

# Calcule el dia correspondiente al 1 del mes y año ingresado
año_bisiesto=False
if año % 4 is 0 and año % 100 is not 0 or año % 400 is 0:
    año_bisiesto = True
else:
    año_bisiesto = False
valor = digitos_año + digitos_año//4
#Cambiando la longitud de febrero cuando es bisiesto
if año_bisiesto:
    longitud_meses[1]=29
#Solo se necesita el primer dia del año
if digitos_siglo is 18:
    valor+=2
elif digitos_siglo is 20:
    valor+=6
if not año_bisiesto:
    valor+=1

primerdia = (valor + 1) % 7
inicio = traductor[primerdia]
valor_dia = 1

#Poblando la lista de meses
for longitud,mes in zip(longitud_meses,lista_meses):
    valor_dia=1
    for semana in mes:
        for dia in semana[inicio:]:
            if valor_dia<=longitud:
                semana[inicio]=valor_dia
                valor_dia+=1
                if inicio<6:
                    inicio+=1
                else:
                    inicio=0
# Muestre el calendario (Mes, Año, Encabezado de dias y la distribución de los dias)
#(Repita estos dos procesos para los 12 meses del año)
for nombre,mes in zip(nombres_meses,lista_meses):
    print(f"\n{nombre} 2019\n")
    print(f"Dom    Lun     Mar    Mie    Jue    Vie    Sab")
    for semana in mes:
        print("{:3}    {:3}    {:3}    {:3}    {:3}    {:3}    {:3}".format(*semana))
```
