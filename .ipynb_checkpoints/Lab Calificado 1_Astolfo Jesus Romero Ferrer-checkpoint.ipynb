## Problema 1 (Puntos 12)
Diseñe e implemente un sistema de Registro de Alummnos. Los datos de los alumnos se encuentran registrados en un archivo alummos.json que contiene la información de cada alumno registrado. Revise el archivo para conocer los campos registrados.

Debe de volcar ese archivo json hacia una lista de alumnos que contendrá una secuencia de objetos clase Alumno. Debe de crear una clase Alumno que permita registrar cada propiedad de cada alumno (los atributos de cada alumno)

Con esta información, debe de diseñar e implementar una interface gráfica de usuario que permita seleccionar de una lista de alumno un registro específico para mostrar los detalles

DISEÑO DE LA INTERFACE GRAFICA:

![](problema1_1.jpg)

RUBRICA DE CALIFICACION:

| DESCRIPCIÓN DE LA TAREA                          | PUNTOS |
|--------------------------------------------------|--------|
| Definicion de la clase Paciente                  | 1/1      |
| Lectura de JSON y creacion de lista de pacientes | 1/1      |
| Diseño de GUI en tkinter                         | 5/6      |
| Uso del Combobox y muestra de datos              | 4/4      |

```python
from tkinter import *
import tkinter.ttk as ttk
import json


class Alumno():
    def __init__(self, nombre='',apellido='',nota1=0,nota2=0,nota3=0,nota4=0,nota5=0):
        self.nombre = nombre
        self.apellido = apellido
        self.nota1 =nota1
        self.nota2 = nota2
        self.nota3 = nota3
        self.nota4 = nota4
        self.nota5 = nota5
        
with open('alumnos.json') as file:
    dic_alumn = json.load(file)
    lista_dic = dic_alumn['alumnos']
    lista_alumn = []
    for dic in lista_dic:
        datos_alumno=list(dic.values())
        lista_alumn.append(Alumno(datos_alumno[0],datos_alumno[1],datos_alumno[2],datos_alumno[3],datos_alumno[4],
                                 datos_alumno[5],datos_alumno[6]))
lista_combo=[]
for alumno in lista_alumn:
    lista_combo.append(alumno.nombre+" "+alumno.apellido)
default=lista_alumn[0].nombre+" "+lista_alumn[0].apellido


    
root = Tk()


alumno_seleccionado = StringVar(value=lista_alumn[0].nombre+" "+lista_alumn[0].apellido)
dic_alumn = dict(zip(lista_combo,lista_alumn))

nombre=StringVar(value=dic_alumn[alumno_seleccionado.get()].nombre)
apellido=StringVar(value=dic_alumn[alumno_seleccionado.get()].apellido)
nota1=StringVar(value=dic_alumn[alumno_seleccionado.get()].nota1)
nota2=StringVar(value=dic_alumn[alumno_seleccionado.get()].nota2)
nota3=StringVar(value=dic_alumn[alumno_seleccionado.get()].nota3)
nota4=StringVar(value=dic_alumn[alumno_seleccionado.get()].nota4)
nota5=StringVar(value=dic_alumn[alumno_seleccionado.get()].nota5)

def actualizar_valores(event):
    global nombre,apellido,nota1,nota2,nota3,nota4,nota5
    global alumno_seleccionado
    global dic_alumn
    print("hola")
    alumno = dic_alumn[alumno_seleccionado.get()]
    nombre.set(alumno.nombre)
    apellido.set(alumno.apellido)
    nota1.set(alumno.nota1)
    nota2.set(alumno.nota2)
    nota3.set(alumno.nota3)
    nota4.set(alumno.nota4)
    nota5.set(alumno.nota5)
    

frm1 = Frame(root)

frm2 = Frame(root)

frm1.pack(side=LEFT,pady=100)
frm2.pack(side=RIGHT,padx=10)

lblAlumnos = Label(frm1,text="Alumnos")
cmbAlumnos= ttk.Combobox(frm1,values=lista_combo,textvariable=alumno_seleccionado)
lblDatos = Label(frm2, text="Datos de Alumno:")
lblNombre = Label(frm2,text=" Nombre ")
lblApellido=Label(frm2,text=" Apellido ")
lblNota1=Label(frm2,text=" Nota1 ")
lblNota2=Label(frm2,text=" Nota2 ")
lblNota3=Label(frm2,text=" Nota3 ")
lblNota4=Label(frm2,text=" Nota4 ")
lblNota5=Label(frm2,text=" Nota5 ")
inNombre = Entry(frm2,textvariable=nombre)
inApellido = Entry(frm2,textvariable=apellido)
inNota1 = Entry(frm2 , textvariable=nota1)
inNota2=Entry(frm2 , textvariable=nota2)
inNota3=Entry(frm2 , textvariable=nota3)
inNota4=Entry(frm2 , textvariable=nota4)
inNota5=Entry(frm2 , textvariable=nota5)

lblAlumnos.grid(column=0,row=0)
cmbAlumnos.grid(column=0,row=1,)


lblDatos.grid(column=0,row=0)
lblNombre.grid(column=0,row=1,sticky=W)
lblApellido.grid(column=0,row=2,sticky=W)
lblNota1.grid(column=0,row=3,sticky=W)
lblNota2.grid(column=0,row=4,sticky=W)
lblNota3.grid(column=0,row=5,sticky=W)
lblNota4.grid(column=0,row=6,sticky=W)
lblNota5.grid(column=0,row=7,sticky=W)

inNombre.grid(column=1,row=1,pady=5,sticky=E)
inApellido.grid(column=1,row=2,pady=5)
inNota1.grid(column=1,row=3,pady=5)
inNota2.grid(column=1,row=4,pady=5)
inNota3.grid(column=1,row=5,pady=5)
inNota4.grid(column=1,row=6,pady=5)
inNota5.grid(column=1,row=7,pady=5)

cmbAlumnos.bind("<<ComboboxSelected>>", actualizar_valores)

root.mainloop()
```

## Problema 2 (8 puntos)
Diseñe e implemente un programa que muestre de forma interactiva una señal senoidal con ampitud y frecuencia variable dependiendo de unos controles gráficos.

La señal senoidal debe mostrarse en un marco tempral de 1 segundo. La amplitdud tendrá un rango de 0.0 y 5.0 voltios (su variacion será en pasos de 0.1 voltios). La frecuencia estará en el rango de 1 a 25 Hx (su variación será en pasos de 1 Hz). Sus controles no deben de salir de este rango.

Los valores iniciales serán amplitud = 1V y frecuencia = 1 Hz. Muestre en el titulo del gráfico ambos datos

DISEÑO DE LA INTERFACE GRAFICA:

![](problema2.jpg)

RUBRICA DE CALIFICACION:

| DESCRIPCIÓN DE LA TAREA                          | PUNTOS |
|--------------------------------------------------|--------|
| Diseño de GUI en tkinter                         | 4/4      |
| Grafica de la señal y control de la visualizacion| 4/4      |

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from tkinter import *

root = Tk()

frm1= Frame(root)
frm2= Frame(root)
frm21 = Frame(frm2,borderwidth=2, relief=GROOVE)
frm22 = Frame(frm2,borderwidth=2, relief=GROOVE)

frm1.pack(side='top')
frm2.pack(side='bottom')
frm21.pack(side=LEFT,padx=10,pady=15)
frm22.pack(side=RIGHT)

fig = Figure()      #facecolor='black'

ax = fig.add_subplot(111)

graph = FigureCanvasTkAgg(fig, master=frm1)
graph.get_tk_widget().pack(side='top', fill='both', expand=True)

amp = DoubleVar(value=1.0)
frec = IntVar(value=5)


t=np.linspace(0,1,1000)
y=amp.get()*np.sin(2*np.pi*frec.get()*t)

def inc_frec():
    global frec,amp
    if frec.get() == 25:
        pass
    elif frec.get() < 25:
        frec.set(frec.get()+1)
    ax.cla()
    y=amp.get()*np.sin(2*np.pi*frec.get()*t)
    ax.plot(t,y)
    ax.set_title("Onda Senoidal de A={:.1f} V y f= {} Hz".format(amp.get(),frec.get()))
    ax.set_ylim(-5,5)
    ax.grid()
    ax.set_xlim(0,1)
    graph.draw()
    
def inc_amp():
    global frec,amp
    if amp.get() == 5:
        pass
    elif amp.get() < 5:
        amp.set(amp.get()+0.1)
    ax.cla()    
    y=amp.get()*np.sin(2*np.pi*frec.get()*t)
    ax.plot(t,y)
    ax.set_title("Onda Senoidal de A={:.1f} V y f= {} Hz".format(amp.get(),frec.get()))
    ax.set_ylim(-5,5)
    ax.grid()
    ax.set_xlim(0,1)
    graph.draw()

def dec_frec():
    global frec,amp
    if frec.get() == 1:
        pass
    elif frec.get() > 1:
        frec.set(frec.get()-1)
    ax.cla()
    y=amp.get()*np.sin(2*np.pi*frec.get()*t)    
    ax.plot(t,y)
    ax.set_title("Onda Senoidal de A={:.1f} V y f= {} Hz".format(amp.get(),frec.get()))
    ax.set_ylim(-5,5)
    ax.grid()
    ax.set_xlim(0,1)
    graph.draw()

def dec_amp():
    global frec,amp
    if amp.get() == 1:
        pass
    elif amp.get() > 1:
        amp.set(amp.get()-0.1)
    ax.cla()
    y=amp.get()*np.sin(2*np.pi*frec.get()*t)    
    ax.plot(t,y)
    ax.set_title("Onda Senoidal de A={:.1f} V y f= {} Hz".format(amp.get(),frec.get()))
    ax.set_ylim(-5,5)
    ax.grid()
    ax.set_xlim(0,1)
    graph.draw()

lblAmplitud = Label(frm21,text="Amplitud",justify="center",width=10)
lblFrecuencia = Label(frm22,text="Frecuencia")
btnIncAmp = Button(frm21,text=">>",command=inc_amp,width=10)
btnDecAmp = Button(frm21,text="<<",command=dec_amp,width=10)
btnIncFrec = Button(frm22,text=">>",command=inc_frec,width=10)
btnDecFrec = Button(frm22,text="<<",command=dec_frec,width=10)

lblAmplitud.pack(side='top')
btnDecAmp.pack(side=LEFT)
btnIncAmp.pack(side=RIGHT)

lblFrecuencia.pack(side='top')
btnDecFrec.pack(side=LEFT)
btnIncFrec.pack(side=RIGHT)
ax.plot(t,y)
ax.set_title("Onda Senoidal de A={:.1f} V y f= {} Hz".format(amp.get(),frec.get()))
ax.set_ylim(-5,5)
ax.grid()
ax.set_xlim(0,1)
root.mainloop()
```
